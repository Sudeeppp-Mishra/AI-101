```python
s = 0
bill = str()
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = int(input("Enter quantity = "))
    total=price*quantity
    bill = bill + f"{name} {price} {quantity} {total}\n"
    s = s+total
    
print("\nall total = ",s)
print(bill)
```

```python
a = "Ram Shyam Hari Sita Gita Ram"
b = input("Enter search value = ")
if b in a:
    print(" Yes there is",b)
    print(a.count(b))
else:
    print("No")
```

```python
#check without case sensitivity
a = "Ram Shyam Hari Sita Gita Ramesh".lower()
b = input("Enter search value = ").lower()
if b in a:
    print(" Yes there is",b)
    print(a.count(b))
else:
    print("No")
```
