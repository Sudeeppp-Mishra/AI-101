# Water Jug Problem

You are given two jugs, a 4 liter one and 3 liter one. None of them have any
measuring marker on it. There is pump that can be used to fill the jugs with water.
How can you get exactly a 2 liter of water into a 4 liter jug ?

Solution:
X = 4 Y = 3
1. (4,0)...................................................................Rule 6
2. (4,0) ..................................................................Rule 1
3. (4,0) → (1,3).....................................................Rule 8
4. (1,3) → (1,0).....................................................Rule 6
5. (1,0)→ (0,1)......................................................Rule 10
6. (0,1)→ (4,1)......................................................Rule 1
7. (4,1)→ (2,3)......................................................Rule 8

```python
print("Water Jug Problem")
x = int(input("Enter X:"))
y = int(input("Enter Y:"))
while True:
    rno = int(input("Enter the Rule No:"))
    # fill bucket with 4L water
    if rno==1:
        if x<4: x=4
    # fill bucket with 3L water
    if rno==2:
        if y<3: y=3
    # empty bucket x
    if rno==5:
        if x>0: x=0
    # empty bucket y
    if rno==6:
        if y>0: y=0
    # fill bucket x with y content, remaining left on y
    if rno==7:
        if x+y>=4 and y>0: x, y=4, y-(4-x)
    # fill bucket y with x content, remaining left on x
    if rno==8:
        if x+y>=3 and x>0: x, y=x-(3-y),3
    # fill bucket x with y contents
    if rno==9:
        if x+y<=4 and y>0: x,y=x+y, 0
    # fill bucket x with y contents
    if rno==10:
        if x+y<=3 and x>0: x,y=0, x+y
    print("X = ", x)
    print("Y = ", y)
    if x==2:
        print("The result is a Goal state")
        break
```

# Breadth First Search

```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

visited = [] # List to keep track of visited nodes
queue = [] # initialize a queue

def bfs(visited, graph, node):
    visited.append(node)
    queue.append(node)
    
    while queue:
        s=queue.pop(0)
        print(s, end=" ")
        
        for neighbour in graph[s]:
            if neighbour not in visited:
                visited.append(neighbour)
                queue.append(neighbour)
                
# driver code
bfs(visited, graph, 'A')
```
